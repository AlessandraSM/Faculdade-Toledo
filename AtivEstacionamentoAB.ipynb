```python
class Node:
    def __init__(self, info):
        self.info = info
        self.proximo = None

    def __repr__(self):
        return f'{self.info} -> {self.proximo}'

    def __str__(self):
        return str(self.info)


class Estaciona:
    def __init__(self):
        self.estacionamento_a = None
        self.estacionamento_b = None
        self.vagas_a = 10
        self.vagas_b = 10
        self._tamanho = 0
        self.primeiro = None
        self.ultimo = None

    def __len__(self):
        return self._tamanho

    def __repr__(self):
        if self._tamanho > 0:
            texto = ""
            aux_ponteiro = self.primeiro
            while aux_ponteiro:
                texto += str(aux_ponteiro.info) + "\n"
                aux_ponteiro = aux_ponteiro.proximo
            return texto
        return "A Fila está vazia!"

    def __str__(self):
        return self.__repr__()

    def estacionar(self, placa, estacionamento):
        novo_no_estacionamento = Node(placa)
        if estacionamento == 'A':
            if self.vagas_a > 0:
                if self.estacionamento_a is None:
                    self.estacionamento_a = novo_no_estacionamento
                    self.primeiro = novo_no_estacionamento
                    self.ultimo = novo_no_estacionamento
                else:
                    self.ultimo.proximo = novo_no_estacionamento
                    self.ultimo = novo_no_estacionamento
                self.vagas_a -= 1
                self._tamanho += 1
                print(f"Veículo {placa} estacionado no Estacionamento A.")
            else:
                print("Estacionamento A indisponível.")
        elif estacionamento == 'B':
            if self.vagas_b > 0:
                if self.estacionamento_b is None:
                    self.estacionamento_b = novo_no_estacionamento
                    self.primeiro = novo_no_estacionamento
                    self.ultimo = novo_no_estacionamento
                else:
                    self.ultimo.proximo = novo_no_estacionamento
                    self.ultimo = novo_no_estacionamento
                self.vagas_b -= 1
                self._tamanho += 1
                print(f"Veículo {placa} estacionado no Estacionamento B.")
            else:
                print("Estacionamento B lotado.")
        else:
            print("Estacionamento inválido.")

    def retirar(self, estacionamento):
        if self._tamanho > 0:
            if estacionamento == 'A':
                if self.estacionamento_a is not None:
                    info = self.estacionamento_a.info
                    self.estacionamento_a = self.estacionamento_a.proximo
            elif estacionamento == 'B':
                if self.estacionamento_b is not None:
                    info = self.estacionamento_b.info
                    self.estacionamento_b = self.estacionamento_b.proximo
            else:
                raise ValueError('Estacionamento inválido.')
            self._tamanho -= 1
            if self._tamanho == 0:
                self.primeiro = None
                self.ultimo = None
            return info
        raise IndexError('A fila está vazia')

    def retirar(self, estacionamento):
        if self._tamanho > 0:
            if estacionamento == 'A':
                if self.estacionamento_a is not None:
                    info = None
                    if self.estacionamento_a.proximo is None:
                        info = self.estacionamento_a.info
                        self.estacionamento_a = None
                        self.primeiro = None
                        self.ultimo = None
                    else:
                        aux_ponteiro = self.estacionamento_a
                        while aux_ponteiro.proximo.proximo is not None:
                            aux_ponteiro = aux_ponteiro.proximo
                        info = aux_ponteiro.proximo.info
                        aux_ponteiro.proximo = None
                        self.ultimo = aux_ponteiro
            elif estacionamento == 'B':
                if self.estacionamento_b is not None:
                    info = self.estacionamento_b.info
                    self.estacionamento_b = self.estacionamento_b.proximo
                    if self.estacionamento_b is None:
                        self.primeiro = None
                        self.ultimo = None
            else:
                raise ValueError('Estacionamento inválido.')

            self._tamanho -= 1
            return info

        raise IndexError('A fila está vazia')


    def carros_estacionados(self, estacionamento):
        placas = []
        if estacionamento == 'A':
            aux_ponteiro = self.estacionamento_a
        elif estacionamento == 'B':
            aux_ponteiro = self.estacionamento_b
        else:
            raise ValueError('Estacionamento inválido.')

        while aux_ponteiro:
            placas.append(aux_ponteiro.info)
            aux_ponteiro = aux_ponteiro.proximo
        print(f"Placas no Estacionamento {estacionamento}")
        return placas


    def encontrar_posicao(self, placa):
        aux_ponteiro = self.estacionamento_a
        index = 0

        while aux_ponteiro:
            if aux_ponteiro.info == placa:
                return index
            aux_ponteiro = aux_ponteiro.proximo
            index += 1

        aux_ponteiro = self.estacionamento_b
        index = 0

        while aux_ponteiro:
            if aux_ponteiro.info == placa:
                return index
            aux_ponteiro = aux_ponteiro.proximo
            index += 1

        raise ValueError(f'A placa {placa} não foi encontrada.')

fila = Estaciona()
fila.estacionar('AAA1000', 'A')
fila.estacionar('AAA1001', 'A')
fila.estacionar('AAA1002', 'A')
fila.estacionar('BBB2001', 'B')
fila.estacionar('BBB2002', 'B')
fila.estacionar('BBB2003', 'B')
print('------------------------')
print(fila.carros_estacionados('A'))
print('------------------------')
print(fila.retirar('A'))
print('------------------------')
print(fila.carros_estacionados('A'))
print('------------------------')
print(fila.carros_estacionados('B'))
print('------------------------')
print(fila.retirar('B'))
print('------------------------')
print(fila.carros_estacionados('B'))
print('-------------------------------------------------------------')
try:
    posicao_a = fila.encontrar_posicao('AAA1000')
    print(f'A placa AAA1000 está na posição {posicao_a} na fila do estacionamento A.')
except ValueError as e:
    print(e)
print('-------------------------------------------------------------')
try:
    posicao_b = fila.encontrar_posicao('BBB2002')
    print(f'A placa BBB2002 está na posição {posicao_b} na fila do estacionamento B.')
except ValueError as e:
    print(e)





```
